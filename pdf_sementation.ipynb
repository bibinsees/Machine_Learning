```python
import PyPDF2

# Define function to extract text from a PDF
def extract_text(pdf_path):
    # Open the PDF file in binary mode
    with open(pdf_path, 'rb') as pdf_file:
        # Create PDF reader object
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        # Initialize empty string to store extracted text
        extracted_text = ''
        # Loop through each page of the PDF
        for page_num in range(len(pdf_reader.pages)):
            # Get the current page object
            page_obj = pdf_reader.pages[page_num]
            # Extract the text from the page
            text = page_obj.extract_text()
            # Add the extracted text to the string
            extracted_text += text
    # Return the extracted text
    return extracted_text

# Define function to segment text
def segment_text(text):
    # Initialize empty list to store segmented text
    segments = []
    # Split the text by newline character
    lines = text.split('\n')
    # Loop through each line of the text
    for line in lines:
        # If the line is not empty, add it to the list of segments
        if line.strip() != '':
            segments.append(line)
    # Return the segmented text
    return segments

pdf_path = '/home/bibin/Downloads/Arbeitsblatt-Aufgaben-Zahlen.pdf'
text = extract_text(pdf_path)
segments = segment_text(text)
print(segments)
```

```python
import PyPDF2

# Define function to extract text from a PDF
def extract_text(pdf_path):
    # Open the PDF file in binary mode
    with open(pdf_path, 'rb') as pdf_file:
        # Create PDF reader object
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        # Initialize empty string to store extracted text
        extracted_text = ''
        # Loop through each page of the PDF
        for page_num in range(len(pdf_reader.pages)):
            # Get the current page object
            page_obj = pdf_reader.pages[page_num]
            # Extract the text from the page
            text = page_obj.extract_text()
            # Add the extracted text to the string
            extracted_text += text
    # Return the extracted text
    return extracted_text

# Define function to segment text
def segment_text(text):
    # Initialize empty list to store segmented text
    segments = []
    # Split the text by newline character
    lines = text.split('\n')
    # Loop through each line of the text
    for line in lines:
        # If the line is not empty, add it to the list of segments
        if line.strip() != '':
            # Check the number of spaces before the text to determine if it is a heading or subheading
            num_spaces = len(line) - len(line.lstrip())
            if num_spaces == 0:
                # This is a heading
                segments.append(line.strip())
            elif num_spaces == 4:
                # This is a subheading
                segments.append(line.strip())
            else:
                # This is a regular line of text
                segments.append(line)
    # Return the segmented text
    return segments

pdf_path = '/home/bibin/Downloads/Arbeitsblatt-Aufgaben-Zahlen.pdf'
text = extract_text(pdf_path)
segments = segment_text(text)
print(segments)
```

```python
import PyPDF2
import spacy

# Load a PDF file
pdf_file = open('/home/bibin/Downloads/beko.pdf', 'rb')
pdf_reader = PyPDF2.PdfReader(pdf_file)

# Initialize a spaCy model for natural language processing
nlp = spacy.load('en_core_web_sm')

# Iterate over the pages in the PDF file
for page_num in range(len(pdf_reader.pages)):
    page = pdf_reader.pages(page_num)
    page_text = page.extractText()

    # Split the page text into lines and remove leading/trailing whitespace
    lines = [line.strip() for line in page_text.split('\n')]

    # Group the lines into sections based on boldness or section numbers
    sections = []
    current_section = {'heading': '', 'subheading': '', 'text': ''}
    for line in lines:
        # Check if the line is a bold heading or subheading
        bold = False
        for token in nlp(line):
            if token.is_stop or token.is_punct:
                continue
            if token.is_upper and token.is_title:
                bold = True
                break

        if bold:
            # line is a bold heading or subheading
            if current_section['text']:
                sections.append(current_section)
            current_section = {'heading': line, 'subheading': '', 'text': ''}
        elif line.strip() and line.strip().startswith(('1.', '2.', '3.', '4.', '5.', '6.', '7.', '8.', '9.')):
            # line starts with a number followed by a period, e.g. 1.1 Introduction
            if current_section['text']:
                sections.append(current_section)
            parts = line.strip().split(' ', maxsplit=1)
            current_section = {'heading': parts[1], 'subheading': parts[0], 'text': ''}
        else:
            # Add the line to the current section text
            current_section['text'] += line + '\n'

    # Add the last section to the list
    if current_section['text']:
        sections.append(current_section)

    # Print the sections for this page
    for section in sections:
        print(f"Heading: {section['heading']}")
        print(f"Subheading: {section['subheading']}")
        print(f"Text: {section['text']}\n")
```
