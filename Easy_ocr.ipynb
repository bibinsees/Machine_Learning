```python
from pdf2image import convert_from_path
import easyocr
import numpy as np
import PIL
from PIL import ImageDraw
import spacy
```

```python
reader = easyocr.Reader(['en'])
```

```python
pdf_path = "/content/beko_removed.pdf"
```

```python
images = convert_from_path(pdf_path)
```

```python
from IPython.display import display, Image
display(images[1])
```

```python
bounds = reader.readtext(np.array(images[1]), min_size=0, slope_ths=0.2, ycenter_ths=0.7, height_ths=0.6, width_ths=0.8, decoder = 'beamsearch',beamWidth = 10)
bounds
```

```python
def draw_boxes(image_to_show,bounds,color='yellow',width =2):
  draw = ImageDraw.Draw(image_to_show)
  for bound in bounds:
    p0,p1,p2,p3 = bound[0]
    draw.line([*p0,*p1,*p2,*p3,*p0], fill=color, width=width)
  return image_to_show
draw_boxes(images[1],bounds)
```

```python
bounds[1][1]
```

```python
text = ''
for i in range(len(bounds)):
  text = text + bounds[i][1] + '\n'
print(text)
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(text)
```

```python
from spacy import displacy
displacy.render(nlp(doc.text),style='ent',jupyter=True)
```
